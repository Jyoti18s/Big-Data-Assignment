```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import time
import pandas as pd

def scrape_amazon(search_keyword, num_pages):
    base_url = 'https://www.amazon.in/s'
    data = []

    for page in range(1, num_pages + 1):
        headers = {'User-Agent': UserAgent().random}
        params = {'k': search_keyword, 'page': page}

        try:
            webpage = requests.get(base_url, params=params, headers=headers)
            webpage.raise_for_status()  # Raise an HTTPError for bad responses

            soup = BeautifulSoup(webpage.content, 'html.parser')

            product_containers = soup.find_all('div', class_='puis-card-container s-card-container s-overflow-hidden aok-relative puis-include-content-margin puis puis-v1x9xtdrqwkf2b2sy3gv3bhvhpn s-latency-cf-section puis-card-border')

            for container in product_containers:
                name = container.find('h2').text.strip()
                rating_element = container.find('span', class_='a-icon-alt')
                rating = rating_element.text.strip() if rating_element else 'N/A'
                reviews_element = container.find('span', class_='a-size-base s-underline-text')
                reviews = reviews_element.text.strip() if reviews_element else 'N/A'
                price_element = container.find('span', class_='a-price-whole')
                price = price_element.text.strip() if price_element else 'N/A'

                data.append({'brand_name': name, 'rating': rating, 'reviews': reviews, 'price': price})

            time.sleep(2)  # Add a delay of 2 seconds between requests to avoid overloading the server

        except requests.exceptions.RequestException as err:
            print("Something went wrong:", err)

    return pd.DataFrame(data)

# Example usage
search_keyword = 'phone'
num_pages_to_scrape = 3
df = scrape_amazon(search_keyword, num_pages_to_scrape)
print(df)
```
