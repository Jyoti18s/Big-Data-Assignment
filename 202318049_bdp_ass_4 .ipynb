##TASK-1:

Generate 100 random numbers in range 0 to 10 using numpy randint function with the seed set to 10. Create a RDD using the parallelize function using data generated in previous step. Calculate the frequency of each number (0 - 10) using appropriate function of RDD.

```python
np.random.seed(10)
ran_num = np.random.randint(0, 11, size=100)
ran_num
```

```python
sc = SparkContext()
rdd = sc.parallelize(ran_num)
```

```python
# frequency calculation
frequency_counts = rdd.map(lambda x: (x, 1)).reduceByKey(lambda x, y: x + y).sortByKey()
frequency_counts_list = frequency_counts.collect()
for number, count in frequency_counts_list:
    print("Number:", number, "-> Count:", count)
sc.stop()
```

##TASK-2:
    
In this task you will calculate the frequency of each word in text8 dataset mentioned above. Create a RDD using the text8 dataset.Use appropriate functions of the RDD to get the word frequencies. Filter the RDD using appropriate function to get the frequencies of words containing the letter ’a’.

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("WordFrequency") \
    .getOrCreate()

sc = spark.sparkContext
```

```python
text8_rdd = sc.textFile("/content/drive/MyDrive/text8")
words_rdd = text8_rdd.flatMap(lambda line: line.split())

# Count the frequency of each word
word_freq = words_rdd.map(lambda word: (word, 1)).reduceByKey(lambda x, y: x + y)

# Filter the RDD to get the frequencies of words containing the letter 'a'
words_with_a_freq = word_freq.filter(lambda x: 'a' in x[0])

# Collect and print the result
print("Frequencies of words containing the letter 'a':")
for word, freq in words_with_a_freq.collect():
    print(f"Word: {word}, Frequency: {freq}")

# Stop SparkContext
sc.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import corr

# Create a SparkSession
spark = SparkSession.builder \
    .appName("Calculate Pearson Correlation") \
    .getOrCreate()

# Read iris JSON data into a Spark DataFrame
iris = spark.read.json("/content/drive/MyDrive/iris.json")

# Calculate Pearson Correlation between petalLength and petalWidth
pearson_corr = iris.select(corr("petalLength", "petalWidth")).collect()[0][0]
print("Pearson correlation between petalLength and petalWidth is :", pearson_corr)
```
